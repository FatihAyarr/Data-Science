### Run this cell before you start to load the necessary libraries and define functions.

```python
import numpy as np
import pandas as pd
from sklearn import model_selection as ms
import sklearn
```

### Open the dataset with Pandas, define the cells with '?' as NaN

```python
df = pd.read_csv('chronic_kidney_disease.csv')
df = df.replace('?', np.nan)
df.head(5)
```

### Write numeric and categorical columns in the dataset as a list. Create the categorical, numeric and target DataFrames.

```python
col_num = ['age','bp','bgr','bu','sc','sod','pot','hemo','pcv','wc','rc'] #list
col_cat = ['sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad',
           'appet', 'pe', 'ane'] #list
col_tar = 'class' #string

df_num = df[col_num]
df_cat = df[col_cat]
y = df[col_tar]
```

### Fill in the missing data (if any) in the numeric DataFrame for each column with the average value of that column. Normalize each column in the DataFrame to min-max within itself.

```python
df[col_num] = df[col_num].astype(float)
df_num = df_num.astype(float)
```

```python
for c in df_num:
    if df_num[c].isna().sum()>0:
        df_num[c].fillna(df_num[c].mean(),inplace=True)
df_num.head(10)
```

```python
df_num2 = df_num.copy()

for i in df_num2:
    count=0
    mn = df_num2[i].min()
    mx = df_num2[i].max()
    for j in df_num2[i]:
        jnew = (j-mn)/(mx-mn)
        df_num2[i][count]=jnew
        count+=1

df_num2.head()
```

### Fill in the missing data (if any) in the Categorical DataFrame for each column with the most frequently repeated value in that column.

```python
for c in df_cat:
    if df_cat[c].isna().sum()>0:
        df_cat[c].fillna(df_cat[c].mode()[0],inplace=True)
df_cat.head()
```

### Apply encoding to the columns in the Categorical DataFrame. You can apply any method you want here. You can replace values like yes/no as 1/0; you can apply one-hot encoding to those that need it. Or you can apply one-hot encoding to all of them.

```python
df2 = df_cat
df2.head()
```

```python
sg_dum = pd.get_dummies(df2)
sg_dum.head()
```

### Combine the Numeric and Categorical DataFrames with axis=1 to create the feature (X) DataFrame.

```python
X = pd.concat([df_num,sg_dum],axis=1)
print(X.isna().sum().sum()) # should be 0
X.head()
```

### Replace the values in y as notckd:0, ckd:1.

```python
vals = {'notckd':0,'ckd':1}
y = y.replace(vals)
y.head() # should only be 0 or 1.
```

### Apply train-test split to x and y with a test rate of 30%.
### Create the data X_train, X_test, y_train, y_test.

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X,y,test_size=0.3,)
```

### Below is a function that predicts with Gaussian Naive Bayes and outputs the results.
### Add confusion matrix to the output of this function. The output should be as follows:
`(y_true,y_pred,y_proba,conf_mat)`

```python
liste = []
liste.clear()
liste
```

```python
#sklearn.naive_bayes.GaussianNB
from sklearn.naive_bayes import GaussianNB

def predict_gnb(X_train,X_test,y_train,y_test):

    clf = GaussianNB()
    clf.fit(X_train,y_train)
    x = clf.score(X_test,y_test)

    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)

    y_true = y_test

    conf_mat1 = sklearn.metrics.confusion_matrix(y_true,y_pred)
    liste.append(conf_mat1)

    return (y_true,y_pred,y_proba,conf_mat1,x)

predict_gnb(X_train,X_test,y_train,y_test)
```

### Prepare the same function (with confusion matrix) for Decision Tree as you modified above:

```python
#sklearn.tree.DecisionTreeClassifier
from sklearn import tree

def predict_dt(X_train,X_test,y_train,y_test):
    clf = tree.DecisionTreeClassifier()
    clf.fit(X_train,y_train)
    x = clf.score(X_test,y_test)

    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)

    y_true = y_test

    conf_mat2 = sklearn.metrics.confusion_matrix(y_true,y_pred)
    liste.append(conf_mat2)

    return (y_true,y_pred,y_proba,conf_mat2,x)

predict_dt(X_train,X_test,y_train,y_test)
```

### Prepare the same function (with confusion matrix) for KNN as you modified above:

```python
#sklearn.neighbors.KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

def predict_knn(X_train,X_test,y_train,y_test,k=1):
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train,y_train)
    x = clf.score(X_test,y_test)

    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)

    y_true = y_test

    conf_mat3 = sklearn.metrics.confusion_matrix(y_true,y_pred)
    liste.append(conf_mat3)

    return (y_true,y_pred,y_proba,conf_mat3,x)

predict_knn(X_train,X_test,y_train,y_test)
```

### Write the function that extracts performance metrics from confusion matrix

```python
def performance_metrics(cm):
    tp = cm[0][0]
    fp = cm[1][0]
    tn = cm[1][1]
    fn = cm[0][1]
    accuracy = (tp + tn) / (tp+fp+tn+fn)
    precision = tp / (tp + fp)
    tpr = tp / (tp + fn)
    fpr = fp / (fp + tn)

    return accuracy, precision, tpr, fpr
```
